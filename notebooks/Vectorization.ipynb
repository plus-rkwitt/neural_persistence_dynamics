# Visuals for the PH vectorizations

Notebook to help understand how the vectorization strategy is parametrized.

```python
%load_ext autoreload
%autoreload 2

%matplotlib inline
import matplotlib.pyplot as plt 

import os
import sys
import copy
import torch
import numpy as np

sys.path.append('../')
from compute_vec import compute_SLayerExponential_parameters
```

```python
def show_vectorization_parameters(pts, centers, num=10000):
    idx = torch.randperm(pts.shape[0])[0:num]
    plt.figure(figsize=(5,5))
    plt.plot(pts[idx,0],pts[idx,1],'.', color='gray')
    plt.plot(centers[:,0],
             centers[:,1],'+',color='blue', markersize=2, label='Centers')
    plt.ylabel('Persistence')
```

Load sample diagrams and make a copy.

```python
raw_dgms_org = torch.load('../data/Giusti23a/1k/dgms_1k_vr_h0h1.pt')
```

Compute vectorization parameters (for $H_0$) and visualize:

```python
raw_dgms0_dup = copy.deepcopy(raw_dgms_org[0])
ci0, si0, pts0 = compute_SLayerExponential_parameters(raw_dgms0_dup, 20, 50000, return_points=True)
show_vectorization_parameters(pts0, ci0)
```

Compute vectorization parameters (for $H_1$) and visualize:

```python
raw_dgms1_dup = copy.deepcopy(raw_dgms_org[1])
ci1, si1, pts1 = compute_SLayerExponential_parameters(raw_dgms1_dup, 20, 50000, return_points=True)
show_vectorization_parameters(pts1, ci1)
```
